### Теория распознавания образов

Теория распознавания образа[1] — раздел информатики и смежных дисциплин, развивающий основы и методы классификации и идентификации предметов, явлений, процессов, сигналов, ситуаций и т. п. объектов, которые характеризуются конечным набором некоторых свойств и признаков. Такие задачи решаются довольно часто, например, при переходе или проезде улицы по сигналам светофора. Распознавание цвета загоревшейся лампы светофора и знание правил дорожного движения позволяет принять правильное решение о том, можно или нельзя переходить улицу.

Необходимость в таком распознавании возникает в самых разных областях — от военного дела и систем безопасности до оцифровки аналоговых сигналов.

Проблема распознавания образа приобрела выдающееся значение в условиях информационных перегрузок, когда человек не справляется с линейно-последовательным пониманием поступающих к нему сообщений, в результате чего его мозг переключается на режим одновременности восприятия и мышления, которому свойственно такое распознавание .

Неслучайно, таким образом, проблема распознавания образа оказалась в поле междисциплинарных исследований — в том числе в связи с работой по созданию искусственного интеллекта, а создание технических систем распознавания образа привлекает к себе всё большее внимание. 

### ResNet: «остаточные» CNN для классификации изображений (Deep Residual Learning for Image Recognition)

#### Проблемы с глубокими нейронными сетями

Эмпирически, глубина сети имеет решающее значение для производительности модели; когда количество слоев сети увеличивается, сеть может выполнять извлечение более сложных моделей признаков, поэтому теоретически лучшие результаты могут быть достигнуты при более глубокой модели. Экспериментально проблема деградации обнаружена в глубоких сетях: когда глубина сети увеличивается, точность сети становится насыщенной и даже снижается. 

Авторы ResNet объясняют явление деградации трудностью достижения постоянного преобразования (y=x) в глубоких нейронных сетях. 

![ResNet struct](https://img-blog.csdnimg.cn/20181119205219323.png)


#### Основная идея для решения проблем
Для структуры с уложенными слоями (несколько слоев уложены друг на друга), когда входным сигналом является x, выученная функция обозначается как H(x), теперь мы хотим, чтобы она выучила остаток F(x) = H(x)-x так, чтобы на самом деле исходная выученная функция была H(x) = F(x)+x. Это происходит потому, что обучение по остаточному признаку проще, чем прямое обучение по исходным признакам. Когда остатки равны 0, уровень стекирования выполняет только постоянное отображение, по крайней мере, производительность сети не снижается. На самом деле остатки не будут равны 0. Это также позволит слою стекирования изучать новые признаки поверх входных признаков и, таким образом, иметь лучшую производительность. Формулировка F(x) + x может быть реализована с помощью нейронных сетей с соединениями для быстрого доступа.

![ResNet struct](https://iq.opengenus.org/content/images/2020/01/resnet.png "Структура соединения для быстрого доступа")

Соединения для быстрого доступа не требуют ни дополнительных параметров, ни сложности вычислений. Это не только привлекательно на практике, но и важно для наших сравнений между простыми и остаточными сетями.

Форма остаточной функции F является переменной. Эксперименты в данной работе включают функции F с двумя или тремя уровнями.

![ResNet struct](https://img-blog.csdnimg.cn/20181119210658243.png)

#### Архитектуры сетей

Начиная с вдохновения Vgg, авторы конструируют простую сверточную нейронную сеть, а затем проектируют остаточную сеть, используя следующие два правила проектирования.

1. При одинаковом размере выходной карты признаков слои имеют одинаковое количество фильтров.
2. Если размер карты признаков уменьшается вдвое, количество фильтров удваивается, чтобы сохранить временную сложность каждого слоя. Мы выполняем понижающую дискретизацию непосредственно через конволюционный слой с размером шага 2.

Последние две сети, представленные ниже, заканчиваются слоем объединения глобальных средних и 1000-мерным полностью подключенным слоем с softmax.

![ResNet struct](https://img-blog.csdnimg.cn/2018111921094853.png)

Основываясь на простой сети, описанной выше, авторы вставляют короткие связи для преобразования сети в соответствующую версию остатков. Если вход и выход имеют одинаковые размеры, можно напрямую использовать постоянное короткое соединение. Когда размерность увеличивается (пунктирные короткие связи на рисунке 3), авторы рассматривают два варианта.
1. короткое соединение по-прежнему выполняет постоянное отображение, с дополнительным нулевым входом для увеличения размерности. Эта опция не вводит дополнительных параметров.
2. Для согласования размеров используется проекционная связь (выполняется сверткой 1 × 1). Для обоих вариантов они выполняются с шагом 2, когда соединение ярлыка охватывает два измерения карты признаков (сетка Fm делается меньше).

![ResNet struct](https://img-blog.csdnimg.cn/20181119211154910.png)

### Оптимизатор SAM

Sharpness-Aware Minimization



#### Проблема обобщающей способности
В современных моделях с чрезмерной параметризацией мало гарантий того, что величина потерь при обучении окажет какое-либо влияние на способность модели к обобщению. На самом деле, оптимизация только значений потерь при обучении, как это часто делается, может легко привести к неоптимальному качеству модели.

#### Основная идея для решения проблем

При градиентном спуске или любом другом алгоритме оптимизации целью является поиск параметра с низким значением потерь. Однако SAM достигает лучшего обобщения, чем другие традиционные методы оптимизации, поскольку фокусируется на поиске параметров с равномерно низкими значениями потерь в пределах области (а не только самих параметров с низкими значениями потерь).

Поскольку вычисляются параметры окрестности, а не отдельные параметры, гиперплоскость потерь более плоская, чем при других методах оптимизации, что, в свою очередь, улучшает обобщение модели.

#### Описание метода

Сначала мы создаем оптимизатор, наследуя класс optimiser от Pytorch, хотя SAM не является новым оптимизатором, а скорее обновляет градиент (с помощью базового оптимизатора) на каждом шаге, где его нужно унаследовать от класса.

Этот класс принимает параметры модели, базовый оптимизатор и rho, который является размером окрестности для вычисления максимальных потерь.

```python
class CifarDataset(Dataset):
     def __init__(self, X, y, transform=None, p=0.0):
         assert X.size(0) == y.size(0)
         super(Dataset, self).__init__()
         self.X = X
         self.y = y
         self.transform = transform
         self.prob = p
        
     def __len__(self):
         return self.y.size(0)
        
     def __getitem__(self, index):
        x = self.X[index]
        if self.transform and np.random.random()<self.prob:
            x = self.transform(x.permute(2, 0, 1)/255.).permute(1, 2, 0)*255.
        y = self.y[index]
        return x, y

transform = T.Compose([
     T.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.2, hue=0.0),
     T.RandomAffine(degrees=15, translate=(0.1, 0.1), scale=(0.8, 1.2),
                    shear=5),
])
```

## Создание Pytorch модели

```python
class Normalize(nn.Module):
    def __init__(self, mean, std):
        super(Normalize, self).__init__()
        self.mean = torch.tensor(mean).to(device)
        self.std = torch.tensor(std).to(device)

    def forward(self, input):
        x = input / 255.0
        x = x - self.mean
        x = x / self.std
        return x.permute(0, 3, 1, 2) # nhwc -> nm

class CNN18(nn.Module):
    def __init__(self, classes=100):
        super(CNN18, self).__init__()
        # https://blog.jovian.ai/image-classification-of-cifar100-dataset-using-pytorch-8b7145242df1
        self.seq = nn.Sequential(
            Normalize([0.5074,0.4867,0.4411],[0.2011,0.1987,0.2025]),

            nn.Conv2d(3, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False),
            nn.BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=3, stride=2, padding=1, dilation=1, ceil_mode=False),

            # layer1
            # BasicBlock
            nn.Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            # BasicBlock
            nn.Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(),
            nn.Conv2d(64, 64, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(64, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),

            # layer2
            # BasicBlock
            nn.Conv2d(64, 128, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            nn.BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            # downsample
            nn.Conv2d(64, 128, kernel_size=(1, 1), stride=(2, 2), bias=False),
            nn.BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            # BasicBlock
            nn.Conv2d(128, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(128, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),

            # layer3
            # BasicBlock
            nn.Conv2d(128, 256, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            nn.BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(),
            nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            # downsample
            nn.Conv2d(128, 256, kernel_size=(1, 1), stride=(2, 2), bias=False),
            nn.BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            # BasicBlock
            nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(256, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),

            # layer4
            # BasicBlock
            nn.Conv2d(256, 512, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False),
            nn.BatchNorm2d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            # downsample
            nn.Conv2d(256, 512, kernel_size=(1, 1), stride=(2, 2), bias=False),
            nn.BatchNorm2d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            # BasicBlock
            nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 512, kernel_size=(3, 3), stride=(1, 1), padding=(1, 1), bias=False),
            nn.BatchNorm2d(512, eps=1e-05, momentum=0.1, affine=True, track_running_stats=True),

            nn.AdaptiveAvgPool2d(output_size=(1, 1)),
            nn.Linear(in_features=512, out_features=1000, bias=True)
        )
        self.cl = nn.Linear(in_features=1000, out_features=classes, bias=True)

    def forward(self, input):
        x = self.seq(input)
        return self.cl(x)
```

#### (Adaptive) SAM Optimizer [5] davda54/sam 

This is an **unofficial** repository for [Sharpness-Aware Minimization for Efficiently Improving Generalization](https://arxiv.org/abs/2010.01412) and [ASAM: Adaptive Sharpness-Aware Minimization for Scale-Invariant Learning of Deep Neural Networks](https://arxiv.org/abs/2102.11600). Implementation-wise, SAM class is a light wrapper that computes the regularized "sharpness-aware" gradient, which is used by the underlying optimizer (such as SGD with momentum). This repository also includes a simple [WRN for Cifar10](example); as a proof-of-concept, it beats the performance of SGD with momentum on this dataset.

```python
from sam import SAM
base_optimizer = torch.optim.SGD  # define an optimizer for the "sharpness-aware" update
#optimizer = SAM(model.parameters(), base_optimizer, lr=0.1, momentum=0.9)
```

```python

names = ['Vgg19',
          'ResNet18',
          'ResNet18 with SAM',
          'ResNet34',
          'ResNet50']

# CNN
models = [Vgg19(classes=CLASS_NUM),
          ResNet18(classes=CLASS_NUM),
          ResNet18(classes=CLASS_NUM),
          ResNet34(classes=CLASS_NUM),
          ResNet50(classes=CLASS_NUM)]

# добавляем сглаживание целевых меток, это увеличит значение функции потерь
# но полученная модель будет более устойчивой к выбросам в обучающей выборке
criterions = [nn.CrossEntropyLoss(label_smoothing=0.2),
              nn.CrossEntropyLoss(label_smoothing=0.2),
              nn.CrossEntropyLoss(label_smoothing=0.2),
              nn.CrossEntropyLoss(label_smoothing=0.2),
              nn.CrossEntropyLoss(label_smoothing=0.2)]
# используется SGD c momentum и L2-регуляризацией весов
optimizers = [optim.SGD(models[0].parameters(), lr=3e-2, momentum=0.9,weight_decay=1e-5),
              optim.SGD(models[1].parameters(), lr=3e-2, momentum=0.9,weight_decay=1e-5),
              SAM(models[2].parameters(), base_optimizer, lr=3e-2, momentum=0.9, weight_decay=1e-5),
              optim.SGD(models[3].parameters(), lr=3e-2, momentum=0.9,weight_decay=1e-5),
              optim.SGD(models[4].parameters(), lr=3e-2, momentum=0.9,weight_decay=1e-5)]
# добавляем постепенное уменьшение шага обучения каждые 200 эпох
schedulers = [optim.lr_scheduler.StepLR(optimizers[0], step_size=200, gamma=0.5),
              optim.lr_scheduler.StepLR(optimizers[1], step_size=200, gamma=0.5),
              optim.lr_scheduler.StepLR(optimizers[2], step_size=200, gamma=0.5),
              optim.lr_scheduler.StepLR(optimizers[3], step_size=200, gamma=0.5),
              optim.lr_scheduler.StepLR(optimizers[4], step_size=200, gamma=0.5)]
```

## Обучение модели по эпохам

```python
EPOCHS = 250
REDRAW_EVERY = 20
steps_per_epoch = len(dataloader['train'])
steps_per_epoch_val = len(dataloader['test'])
# NEW
losses_models = []
losses_models_val = []
for name,model,criterion,optimizer,scheduler in zip(names,models,criterions,optimizers,schedulers):
    pbar = tqdm(total=EPOCHS*steps_per_epoch)
    losses = []
    losses_val = []
    passed = 0
    # для создания чекпоинта
    best_acc = 0
    checkpoint_path = name+'_cifar_cnn_aug.pth'
    model = model.to(device)
    for epoch in range(EPOCHS):  # проход по набору данных несколько раз
        tmp = []
        model.train()
        for i, batch in enumerate(dataloader['train'], 0):
            # получение одного минибатча; batch это двуэлементный список из [inputs, labels]
            inputs, labels = batch
            # на GPU
            inputs, labels = inputs.to(device), labels.to(device)
            # очищение прошлых градиентов с прошлой итерации
            optimizer.zero_grad()

            # прямой + обратный проходы + оптимизация
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            if name.find('SAM') == -1:
                loss.backward()
                optimizer.step()
            else:
                # first forward-backward pass
                loss.backward()
                optimizer.first_step(zero_grad=True)
                # second forward-backward pass
                outputs = model(inputs)
                criterion(outputs, labels).backward()  # make sure to do a full forward pass
                optimizer.second_step(zero_grad=True)
            # для подсчёта статистик
            accuracy = (labels.detach().argmax(dim=-1)==outputs.detach().argmax(dim=-1)).\
                        to(torch.float32).mean().cpu()*100
            tmp.append((loss.item(), accuracy.item()))
            pbar.update(1)

        losses.append((np.mean(tmp, axis=0),
                      np.percentile(tmp, 25, axis=0),
                      np.percentile(tmp, 75, axis=0)))
        scheduler.step() # обновляем learning_rate каждую эпоху
        tmp = []
        model.eval()
        with torch.no_grad(): # отключение автоматического дифференцирования
            for i, data in enumerate(dataloader['test'], 0):
                inputs, labels = data
                # на GPU
                inputs, labels = inputs.to(device), labels.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, labels)
                accuracy = (labels.argmax(dim=-1)==outputs.argmax(dim=-1)).\
                            to(torch.float32).mean().cpu()*100
                tmp.append((loss.item(), accuracy.item()))
        losses_val.append((np.mean(tmp, axis=0),
                          np.percentile(tmp, 25, axis=0),
                          np.percentile(tmp, 75, axis=0)))

        # сохранение чекпоинта
        acc = losses_val[-1][0][1]
        if acc > best_acc:
            best_acc = acc
            torch.save(model.state_dict(), checkpoint_path)
        # обновление графиков
        if (epoch+1) % REDRAW_EVERY != 0:
            continue
        clear_output(wait=False)
        print(name+':')
        print('Эпоха: %s\n'
              'Лучшая доля правильных ответов: %s\n'
              'Текущая доля правильных ответов: %s' % (epoch+1, best_acc, acc))
        passed += pbar.format_dict['elapsed']
        pbar = tqdm(total=EPOCHS*steps_per_epoch, miniters=5)
        pbar.update((epoch+1)*steps_per_epoch)
        x_vals = np.arange(epoch+1)
        _, ax = plt.subplots(1, 2, figsize=(15, 5))
        stats = np.array(losses)
        stats_val = np.array(losses_val)
        ax[1].set_ylim(stats_val[:, 0, 1].min()-5, 100)
        ax[1].grid(axis='y')
        for i, title in enumerate(['CCE', 'Accuracy']):
            ax[i].plot(x_vals, stats[:, 0, i], label='train')
            ax[i].fill_between(x_vals, stats[:, 1, i],
                              stats[:, 2, i], alpha=0.4)
            ax[i].plot(x_vals, stats_val[:, 0, i], label='val')
            ax[i].fill_between(x_vals,
                              stats_val[:, 1, i],
                              stats_val[:, 2, i], alpha=0.4)
            ax[i].legend()
            ax[i].set_title(name + title)
        plt.show()
    model.load_state_dict(torch.load(checkpoint_path))
    print('Обучение закончено за %s секунд' % passed)
    losses_models.append(losses)
    losses_models_val.append(losses_val)
```
